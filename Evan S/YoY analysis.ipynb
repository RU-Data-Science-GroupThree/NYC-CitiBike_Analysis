```python
import pandas as pd
```

```python
March_2014 = pd.read_csv("../data/201403-citibike-tripdata.csv")
March_2015 = pd.read_csv("../data/201503-citibike-tripdata.csv")
March_2016 = pd.read_csv("../data/201603-citibike-tripdata.csv")
March_2017 = pd.read_csv("../data/201703-citibike-tripdata.csv")
March_2018 = pd.read_csv("../data/201803-citibike-tripdata.csv")
March_2019 = pd.read_csv("../data/201903-citibike-tripdata.csv")
March_Agg = pd.DataFrame()

myseries = pd.DataFrame(pd.Series([len(March_2014), len(March_2015), len(March_2016), len(March_2017), len(March_2018), len(March_2019)]))
myseries
```

```python
March_2017.columns = ['tripduration', 'starttime', 'stoptime', 'start station id',
       'start station name', 'start station latitude',
       'start station longitude', 'end station id', 'end station name',
       'end station latitude', 'end station longitude', 'bikeid', 'usertype',
       'birth year', 'gender']
```

```python
March_Agg = March_Agg.append([March_2014, March_2015, March_2016, March_2017, March_2018, March_2019], ignore_index = True)
March_Agg
```

```python
#March_Agg['Date'] = [d.date() for d in pd.to_datetime(March_Agg['starttime'],infer_datetime_format = True)]
March_Agg['Date'] = pd.to_datetime(March_Agg['starttime'],infer_datetime_format = True)
March_Agg['Month'] = March_Agg['Date'].dt.month
March_Agg['Year'] = March_Agg['Date'].dt.year
March_Agg['Time'] = March_Agg['Date'].dt.time
March_Agg
```

```python
March_Record_Series = March_Agg[["Date","Month","Year","Time"]]
March_Record_Count = March_Record_Series.groupby(["Year","Month"]).count()
March_Record_Count
```
