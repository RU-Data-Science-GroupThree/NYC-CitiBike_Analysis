```python
import pandas as pd
from datetime import date
```

```python
Jan_19 = pd.read_csv("../data/201901-citibike-tripdata.csv")
Feb_19 = pd.read_csv("../data/201902-citibike-tripdata.csv") 
Mar_19 = pd.read_csv("../data/201903-citibike-tripdata.csv") 
Apr_19 = pd.read_csv("../data/201904-citibike-tripdata.csv") 
May_19 = pd.read_csv("../data/201905-citibike-tripdata.csv") 
Jun_19 = pd.read_csv("../data/201906-citibike-tripdata.csv") 
Jul_19 = pd.read_csv("../data/201907-citibike-tripdata.csv") 
Aug_19 = pd.read_csv("../data/201908-citibike-tripdata.csv") 
Sep_19 = pd.read_csv("../data/201909-citibike-tripdata.csv") 
Oct_19 = pd.read_csv("../data/201910-citibike-tripdata.csv") 
Nov_19 = pd.read_csv("../data/201911-citibike-tripdata.csv") 
Dec_19 = pd.read_csv("../data/201912-citibike-tripdata.csv")
```

```python
Year_Agg = pd.concat([Jan_19, Feb_19, Mar_19, Apr_19, May_19, Jun_19, Jul_19, Aug_19, Sep_19, Oct_19, Nov_19, Dec_19], ignore_index = True)
Year_Agg = Year_Agg.loc[Year_Agg['tripduration']]
Year_Agg
```

```python
Year_Agg['Age'] = date.today().year - Year_Agg['birth year']
Year_Agg
```

```python
youngest = min(Year_Agg['Age'])
oldest = max(Year_Agg['Age'])
bin_list = [youngest, 25, 35, 45, 55, 65, oldest]
my_group_names = ["17-24","25-34","35-44","45-54","55-64","65+"]
Year_Agg["Age Group"] = pd.cut(Year_Agg["Age"],bin_list, labels = my_group_names, include_lowest = True)
Year_Agg
```

```python
DemographicTable = Year_Agg.groupby(["usertype", "gender","Age Group"])["tripduration"].describe()
DemographicTable
```
